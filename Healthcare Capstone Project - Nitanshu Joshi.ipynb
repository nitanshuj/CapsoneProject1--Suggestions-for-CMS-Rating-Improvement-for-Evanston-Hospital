###  Business Understanding
The Centers for Medicare & Medicaid Services (CMS), a federal agency based in the US, is responsible for administering the country’s major healthcare programs, including Medicare and Medicaid. CMS collects, analyses the data and produces research reports. It then works to eliminate the instances of fraud and abuse within the healthcare system. The agency aims to provide a healthcare system with better care, access to coverage and improved health services to the citizens. CMS also aims to reduce the overall health costs such as hospital expenses and insurance premiums for the citizens.<br>
The CMS rates hospitals in the US on a scale of 1-5, with the objective of making it easier for patients and consumers to compare the quality of services offered by various hospitals.<br>
The ratings directly influence the consumers’ choice of hospitals and may significantly impact hospitals' revenues. Thus, it is extremely important for hospitals to identify the key factors that affect their ratings so that they can work on improving them.<br><br><br>

## Defining important functions

```python
# Function for Checking NULL Counts and NULL Percentages
# ======================================================
def check_null(df):
    null_val = pd.DataFrame()
    
    Null_Count = pd.DataFrame(df.isnull().sum().sort_values(ascending=False), 
                              columns=['Null_Count'])
    
    Null_Percentage = pd.DataFrame(round((df.isnull().sum()/len(df))*100, 2).sort_values(ascending=False),
                                   columns=['Null_Percentage'])
    
    return pd.concat([Null_Count, Null_Percentage], axis = 1)
# ---------------------------------------------------------------------------------------------------------------------
```

### Logistic Regression without Hyperparameter Tuning with Stratified K-Fold Cross Validation

```python
# Making a copy of the Dataframe for Logistic Regression
df_logit = df.copy()
```

```python
# Resetting the index to default for the Cross Validation Approach
# ----------------------------------------
df_logit.reset_index(level=0, inplace=True)
```

```python
# Defining List for the various Model Evaluation Metrics
# -------------------------------------------------------
logit_accuracies_1 = []
logit_precision_1 = []
logit_recall_1 = []
logit_f1scores_1 = []
```

```python
# Putting independent variable to X
X = df_logit.drop('Hospital overall rating', axis = 1)

# Putting response variable to y
y = df_logit['Hospital overall rating']

print("Shape of X and y respectively are - ",X.shape, y.shape)
```

```python
# Scaling for the input features
# ------------------------------
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
# Defining the Logistic Regression Model Object
# ---------------------------------------------
logit_ovr = LogisticRegression(random_state=0,multi_class='ovr')
```

```python
# Defining the Stratified K-Fold Cross Validation Object 
# ------------------------------------------------------
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```
